# 📧 Spam Email Classification using Scikit-Learn

## Project Overview
Complete end‑to‑end spam detection model.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
url = "https://raw.githubusercontent.com/justmarkham/scikit-learn-videos/master/data/sms.tsv"
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])
df.head()
```

```python
df['label_num'] = df.label.map({'ham':0, 'spam':1})
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label_num'], test_size=0.2, random_state=42)
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)
```

```python
y_pred = model.predict(X_test_tfidf)
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
sample = ["Congratulations! You've won a free iPhone. Click here now!"]
sample_vec = vectorizer.transform(sample)
model.predict(sample_vec)
```
